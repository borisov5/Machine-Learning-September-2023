```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs, make_circles, make_moons

from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
```

```python
attributes, clusters = make_blobs(n_samples=300, centers=3, cluster_std=2.5)
```

```python
plt.scatter(attributes[:, 0], attributes[:, 1], c=clusters)
plt.show()
```

```python
k_means = KMeans(n_clusters=3, init="k_means++")
```

```python
k_means.fit(attributes)
```

```python
k_means.cluster_centers_
```

```python
k_means.inertia_
```

```python
predictions = k_means.predict(attributes)
```

```python
plt.scatter(attributes[:, 0], attributes[:, 1], c=predictions)
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], c='r')
plt.show()
```

```python
inertias = []
for k in range(1, 11):
    km = KMeans(n_clusters=k)
    km.fit(attributes)
    inertias.append(km.inertia_)
```

```python
inertias
```

```python
plt.plot(range(1,11), inertias, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
attributes_circles, clusters_circles = make_circles(noise=0.1, factor=0.3)
```

```python
plt.scatter(attributes_circles[:, 0], attributes_circles[:, 1], c=clusters_circles)
plt.gca().set_aspect("equal")
plt.show()
```

```python
k_means_circles = KMeans(n_clusters=10, init="k-means++")
k_means_circles.fit(attributes_circles)
predictions_circles = k_means_circles.predict(attributes_circles)
```

```python
plt.scatter(attributes_circles[:, 0], attributes_circles[:, 1], c=predictions_circles)
plt.gca().set_aspect("equal")
plt.scatter(k_means_circles.cluster_centers_[:, 0], k_means_circles.cluster_centers_[:, 1], c='r')
plt.show()
```

```python
inertias = []
for k in range(1, 16):
    km = KMeans(n_clusters=k)
    km.fit(attributes_circles)
    inertias.append(km.inertia_)
```

```python
plt.plot(range(1,16), inertias, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
agglomerative_clustering.labels_
```

```python
plt.scatter(attributes[:, 0], attributes[:, 1], c=agglomerative_clustering.labels_)
plt.show()
```
